#### To check the age distribution by Industry

```python
plt.figure(figsize=(15,10))
sns.boxplot(x= 'Age',  y='Industry', data=df, showmeans = True)
plt.grid()
plt.show()
```

#### Use Strip Plot

Plot a strip plot to check the relationship between the variables 'Rank' and 'Age'


```python
len(df.Country.unique())
```

```python
# strip plot
plt.figure(figsize=(30,10))
sns.swarmplot(x = 'Industry', y = 'Age', data = df)

# display the plot
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.violinplot(x = 'Industry', y = 'Age', data = df)

# display the plot
plt.show()
```
